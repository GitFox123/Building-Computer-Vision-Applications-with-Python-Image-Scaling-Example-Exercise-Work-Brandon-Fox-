```python
# Task #1: Print an image
#
# Copy an image file of your own in this folder, 
# open it and display it.

# Write your code here
img = plt.imread("Onimai_Nemu.jpg")
plt.imshow(img)
```

```python
# Task #2: Downscale that image.
#
# In this cell, downscale your image to the resolution of your choice.
# Feel free to use the downscale() function, or any other function
# either made by yourself, or from a library, like cv2.resize()

# Write your code here

#Let's just use the "downscale()" function here.
downscaledImg = downscale(img,5)
downscaledImg = adapt_image(downscaledImg)
plt.imshow(downscaledImg)
```

```python
# Task #3: Upscale that image.
#
# In this cell, upscale your image to the resolution of your choice.
# Feel free to use the upscale_by_2() function, or any other function
# either made by yourself, or from a library, like cv2.resize()
#

# Write your code here

#Let's try the cv2.resize() function (or really, the cv.resize() function, since the alias of "cv2" is "cv" here!)
#for this task.

#For horizontal stretching.
scalarX = 500
#For vertical stretching.
scalarY = 175
rows = int(downscaledImg.shape[0] * scalarX / 100)
columns = int(downscaledImg.shape[1] * scalarY / 100)
dim = (rows,columns)

print("Row and Column Dimensions: " + str(dim))

upscaledImg = cv.resize(downscaledImg, dim, interpolation = cv.INTER_LINEAR)
upscaledImg = adapt_image(upscaledImg)

plt.imshow(upscaledImg)
#We end up with something blurry because when we downscaled the image, the pixels ended up getting lost in the process.
```
